```python
stack = []
lines = []
while True:
    line = input()
    if line and line != "exit":
        lines.append(line)
    elif line == "exit":
      lines.append(line)
      break
    else:
      break
request = '\n'.join(lines)
```

```python
for line in request.splitlines():
  list_request = line.split()
  if (list_request[0] == 'push'):
    stack.append(list_request[1])
    print("ok")
  elif (list_request[0] == 'pop'):
    if (len(stack) == 0):
      print("error")
    else:
      x = stack.pop()
      print(x)
  elif (list_request[0] == 'back'):
    if (len(stack) == 0):
      print("error")
    else:
      print(stack[-1])
  elif (list_request[0] == 'size'): 
    print(len(stack))
  elif (list_request[0] == 'clear'):
    stack.clear()
    print("ok")
  elif (list_request[0] == 'exit'):
    # flag = 0
    print("bye") 
    break 
```
