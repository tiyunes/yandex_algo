```python
def transformToPrefix(infix):
  stack = []
  ans = []
  for operand in infix:
    if (operand != "+" and operand != "-" and operand != "*" and operand != "(" and operand != ")"):
      ans.append(operand)
    elif (operand == "("):
      stack.append(operand)
    elif (operand == ")"):
      while (stack[-1] != "("):
        if (len(stack) > 0):
           ans.append(stack.pop())
      if (stack[-1] == "("):
        stack.pop()
    elif (operand == "+" or operand == "-"):
      ans.append(stack[-1])
      stack.append(operand)
    elif (operand == "*"):
      while (stack[-1] == "*"):
        ans.append(stack.pop())
      stack.append(operand)
  return ans
```
