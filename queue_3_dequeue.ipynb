```python
lines = []
while True:
    line = input()
    if line and line != "exit":
        lines.append(line)
    elif line == "exit":
      lines.append(line)
      break
    else:
      break
request = '\n'.join(lines)
```

```python
dequeue = []
for line in request.splitlines():
  list_request = line.split()
  if (list_request[0] == 'push_front'):
    dequeue.insert(0, list_request[1])
    print("ok")
  elif (list_request[0] == 'push_back'):
    dequeue.append(list_request[1])
    print("ok")
    # if (len(dequeue) == 0):
    #   print("error")
    # else:
    #   x = dequeue.pop(0)
    #   print(x)
  elif (list_request[0] == 'pop_front'):
    if (len(dequeue) == 0):
      print("error")
    else:
      print(dequeue.pop(0))
  elif (list_request[0] == 'pop_back'):
    if (len(dequeue) == 0):
      print("error")
    else:
      print(dequeue.pop(-1))
  elif (list_request[0] == 'front'):
    if (len(dequeue) == 0):
      print("error")
    else:
      print(dequeue[0])
  elif (list_request[0] == 'back'):
    if (len(dequeue) == 0):
      print("error")
    else:
      print(dequeue[-1])
  elif (list_request[0] == 'size'): 
    print(len(dequeue))
  elif (list_request[0] == 'clear'):
    dequeue.clear()
    print("ok")
  elif (list_request[0] == 'exit'):
    # flag = 0
    print("bye") 
    break 
```
