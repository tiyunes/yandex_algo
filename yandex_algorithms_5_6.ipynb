## 5. Красивая строка

```python
n = int(input())
count_letters = []
ans = 0
for i in range(n):
    count_letters.append(int(input()))
for i in range(n - 1):
    ans += min(count_letters[i], (count_letters[i + 1]))
print(ans)
```

## 6. Операционные системы lite

```python
M, N, count = int(input()), int(input()), 0
disk = []
for i in range(N):
    a, b = map(int, input().split())
    disk.append([a, b, True])
    for j in range(i):
        if disk[j][1] >= disk[i][0] and disk[j][0] <= disk[i][1] and disk[j][2] == True:
            disk[j][2] = False
for i in range(len(disk)):
    if disk[i][2]:
        count += 1
print(count)
```
