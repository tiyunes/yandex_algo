```python
sequence = input()
```

```python
stack = []
flag = 1
for character in sequence:
  if (character == '('):
    stack.append('(')
  elif (character == '{'):
    stack.append('{')
  elif (character == '['):
    stack.append('[')

  elif (character == ')'):
    if (len(stack) > 0):
      if (stack[-1] != '('):
        ans = "no"
        flag = 0
        break
      elif (stack[-1] == '('):
        stack.pop()
    else:
      ans = "no"
      flag = 0
      break

  elif (character == ']'):
    if (len(stack) > 0):
      if (stack[-1] != '['):
        ans = "no"
        flag = 0
        break
      elif (stack[-1] == '['):
        stack.pop()
    else:
      ans = "no"
      flag = 0
      break

  elif (character == '}'):
    if (len(stack) > 0):
      if (stack[-1] != '{'):
        ans = "no"
        flag = 0
        break
      elif (stack[-1] == '{'):
        stack.pop()
    else:
      ans = "no"
      flag = 0
      break
  
  # print(stack)

if(len(stack) == 0 and flag == 1):
  ans = "yes" 
else:
  ans = "no"

print(ans)
```
