**№26 Самый дешевый путь**

В каждой клетке прямоугольной таблицы 
N
×
M
 записано некоторое число. Изначально игрок находится в левой верхней клетке. За один ход ему разрешается перемещаться в соседнюю клетку либо вправо, либо вниз (влево и вверх перемещаться запрещено). При проходе через клетку с игрока берут столько килограммов еды, какое число записано в этой клетке (еду берут также за первую и последнюю клетки его пути).
Требуется найти минимальный вес еды в килограммах, отдав которую игрок может попасть в правый нижний угол.

```python
N, M = map(int, input().split())
penalties = [[0] * M] * N
for i in range(N):
  penalties[i] = list(map(int, input().split()))
```

```python
dp = [] * N
```

```python
for i in range(N):
  dp.append([])
```

```python
dp[0].append(penalties[0][0])
# dp[0][0] = penalties[0][0]

for i in range(1, M):
  dp[0].append(dp[0][i - 1] + penalties[0][i])
for i in range(1, N):
  dp[i].append(dp[i - 1][0] + penalties[i][0])
```

```python
for i in range(1, N):
  for j in range(1, M):
    dp[i].append(min(dp[i - 1][j], dp[i][j - 1]) + penalties[i][j])
```

**№27 Вывести маршрут максимальной стоимости**

В левом верхнем углу прямоугольной таблицы размером 
N
×
M
 находится черепашка. В каждой клетке таблицы записано некоторое число. Черепашка может перемещаться вправо или вниз, при этом маршрут черепашки заканчивается в правом нижнем углу таблицы.
Подсчитаем сумму чисел, записанных в клетках, через которую проползла черепашка (включая начальную и конечную клетку). Найдите наибольшее возможное значение этой суммы и маршрут, на котором достигается эта сумма.

```python
N, M = map(int, input().split())
cost = [[0] * M] * N
for i in range(N):
  cost[i] = list(map(int, input().split()))
dp = [] * N
for i in range(N):
  dp.append([])



dp[0].append(cost[0][0])
# dp[0][0] = penalties[0][0]

for i in range(1, M):
  dp[0].append(dp[0][i - 1] + cost[0][i])
for i in range(1, N):
  dp[i].append(dp[i - 1][0] + cost[i][0])

for i in range(1, N):
  for j in range(1, M):
    dp[i].append(max(dp[i - 1][j], dp[i][j - 1]) + cost[i][j])
    
print(dp[N - 1][M - 1])
```

```python
print(dp)
```

```python
ans = []
i = N - 1
j = M - 1

for k in range(M + N - 2):
  if (i >= 1 and j >= 1):
    if ((dp[i][j] == (dp[i][j - 1] + cost[i][j]))):
      ans.append('R')
      j -= 1
    else:
      ans.append('D')
      i -= 1
  elif (i == 0 and j > 0):
    ans.append('R')
    j -= 1
  elif (i > 0 and j == 0):
    ans.append('D')
    i -= 1
  elif (i == 0 and j == 0):
    break

ans.reverse()
  
print(*ans)
```

**№28 Ход конем**

Дана прямоугольная доска N × M (N строк и M столбцов). В левом верхнем углу находится шахматный конь, которого необходимо переместить в правый нижний угол доски. В данной задаче конь может перемещаться на две клетки вниз и одну клетку вправо или на одну клетку вниз и две клетки вправо.

Необходимо определить, сколько существует различных маршрутов, ведущих из левого верхнего в правый нижний угол

```python
N, M = map(int, input().split())

dp = [] * N
for i in range(N):
  dp.append([])

dp[0].append(1)

if (N >= 2):
  dp[1].append(0)

if (N >= 3):
  dp[2].append(0)
  dp[2].append(1)

for i in range(1, M):
  dp[0].append(0)
  if (N >= 2):
    if (i == 2):
      dp[1].append(1)
    else:
      dp[1].append(0)

for i in range(3, N):
  dp[i].append(0)
  dp[i].append(0)

if (N >= 3 and M >= 3):
  for i in range(2, N):
    for j in range(2, M):
      dp[i].append(dp[i - 1][j - 2] + dp[i - 2][j - 1])
```

**№29 Кафе**

Около Петиного университета недавно открылось новое кафе, в котором действует следующая система скидок: при каждой покупке более чем на 100 рублей покупатель получает купон, дающий право на один бесплатный обед (при покупке на сумму 100 рублей и меньше такой купон покупатель не получает).

Однажды Пете на глаза попался прейскурант на ближайшие N дней. Внимательно его изучив, он решил, что будет обедать в этом кафе все N дней, причем каждый день он будет покупать в кафе ровно один обед. Однако стипендия у Пети небольшая, и поэтому он хочет по максимуму использовать предоставляемую систему скидок так, чтобы его суммарные затраты были минимальны. Требуется найти минимально возможную суммарную стоимость обедов и номера дней, в которые Пете следует воспользоваться купонами.

```python
N = int(input())

infinity = 100000

dp = [] * (N + 1)
costs = [] * (N + 1)

for i in range(N + 1):
  dp.append([])

costs.append(0)
dp[0].append(0)
dp[0].append(infinity)
dp[0].append(infinity)
dp[0].append(infinity)
dp[0].append(infinity)
dp[0].append(infinity)

for i in range(1, N + 1):
  costs.append(int(input()))

for i in range(1, N + 1):
  if (costs[i] > 100):
    if(dp[i - 1][1] != infinity):
      if ((dp[i - 1][0] + costs[i]) <= dp[i - 1][1]):
        dp[i].append(dp[i - 1][0] + costs[i])
      else:
        dp[i].append(dp[i - 1][1])
    else:
      dp[i].append(dp[i - 1][0] + costs[i])

    for j in range(1, i + 2):
      if(dp[i-1][j] == infinity and dp[i - 1][j + 1] == infinity and dp[i-1][j - 1] == infinity):
        dp[i].append(infinity)
      elif(dp[i - 1][j + 1] != infinity):
        if ((min(dp[i - 1][j], dp[i - 1][j - 1]) + costs[i]) <= dp[i - 1][j + 1]):
          dp[i].append(min(dp[i - 1][j], dp[i - 1][j - 1]) + costs[i])
        else:
          dp[i].append(dp[i - 1][j + 1])
      else:
        dp[i].append(min(dp[i-1][j], dp[i - 1][j + 1], dp[i-1][j - 1]) + costs[i])
    
    dp[i].append(infinity)
    dp[i].append(infinity)
    dp[i].append(infinity)

  else:
    if(dp[i - 1][1] != infinity):
      if ((dp[i - 1][0] + costs[i]) <= dp[i - 1][1]):
        dp[i].append(dp[i - 1][0] + costs[i])
      else:
        dp[i].append(dp[i - 1][1])
    else:
      dp[i].append(dp[i - 1][0] + costs[i])
      
    for j in range(1, i + 2):
      # print("i: ", i, ", j:", j)
      if(dp[i-1][j] == infinity and dp[i - 1][j + 1] == infinity):
        dp[i].append(infinity)
      elif(dp[i - 1][j + 1] != infinity):
        if ((dp[i - 1][j] + costs[i]) <= dp[i - 1][j + 1]):
          dp[i].append(dp[i - 1][j] + costs[i])
        else:
          dp[i].append(dp[i - 1][j + 1])
      else:
        dp[i].append(min(dp[i-1][j], dp[i - 1][j + 1]) + costs[i])
        
    
    dp[i].append(infinity)
    dp[i].append(infinity)
    dp[i].append(infinity)

ans = dp[N][0]
ans_idx = 0
for i in range(len(dp[N])):
  if (dp[N][i] <= ans):
    ans = dp[N][i]
    ans_idx = i

print(ans)
```
